```python
import pandas as pd
import geopandas as gpd
```

```python
foreclosures = pd.read_csv('../Data/processed/FC_count.csv', index_col=0)
police = pd.read_csv('../Data/processed/PD_count.csv')
complaints = pd.read_csv('../Data/processed/complaint_count.csv')
owners = gpd.read_file('../Data/Master_Building/CH_Only_Clip.shp')
```

```python
owners.head()
```

```python
addresses = []
for ix,row in owners.iterrows():
    address = str(row['parcel_add'])[:-7]
    addresses.append(address)

owners['parcel_add'] = addresses
owners.head()
```

### Foreclosures

```python
foreclosures = foreclosures[['PPN', 'Count']]
foreclosures.columns = ['PPN', 'FC_count']
```

```python
owners.PARCELPIN = owners.PARCELPIN.astype(str)
foreclosures.PPN = foreclosures.PPN.astype(str)
```

```python
len(owners)
```

```python
featureSpace = pd.merge(owners, foreclosures, how='left', left_on='PARCELPIN', right_on='PPN')
```

```python
len(featureSpace)
```

```python
featureSpace.drop(['PPN'], axis=1, inplace=True)
featureSpace.head()
```

### Police Dispatch

```python
police['address1'] = map(lambda x: x.upper(), police['address1'])
police['address1'] = police['address1'] + ', CLEVELAND HEIGHTS, OH'
police.head()
```

```python
len(featureSpace)
```

```python
featureSpace = pd.merge(featureSpace, police, how='left', left_on='parcel_add', right_on='address1')
featureSpace.drop('address1', axis=1, inplace=True)
len(featureSpace)
```

```python
featureSpace.head()
```

### Complaints

```python
addresses = []
for ix,row in complaints.iterrows():
    address = row['Address'].split(',')[0] + ', Cleveland Heights, OH'
    addresses.append(address)

complaints.Address = addresses
complaints['Address'] = map(lambda x: x.upper(), complaints['Address'])
complaints.head()
```

```python
len(featureSpace)
```

```python
featureSpace = pd.merge(featureSpace, complaints, how='left', left_on='parcel_add', right_on='Address')
featureSpace.drop('Address', axis=1, inplace=True)
#featureSpace.drop_duplicates(inplace=True)
featureSpace.dropna(subset=['geometry'], inplace=True)
len(featureSpace)
```

```python
featureSpace_shp = gpd.GeoDataFrame(featureSpace, crs = {'init': 'epsg:4326'})
featureSpace_shp.crs
```
